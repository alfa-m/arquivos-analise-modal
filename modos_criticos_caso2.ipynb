```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
modos_criticos_df = pd.read_csv("./modos_criticos_y_caso2.csv", index_col=0)
```

```python
modos_criticos_df.sort_values(by="frequência", inplace=True)
modos_criticos_df
```

```python
modos_criticos_df2 = modos_criticos_df.groupby("local").count()
modos_criticos_df2.sort_values(by="frequência", ascending=False, inplace=True)
modos_criticos_df2.describe()
```

```python
modos_criticos_df2
```

```python
fig, ax = plt.subplots()
bar_container = ax.barh(modos_criticos_df2.index, modos_criticos_df2["frequência"], height=0.6)
ax.set(title="Incidência de modos críticos por barra",xlim=(0, 1010))
ax.bar_label(bar_container, padding=2)
plt.show()
```
